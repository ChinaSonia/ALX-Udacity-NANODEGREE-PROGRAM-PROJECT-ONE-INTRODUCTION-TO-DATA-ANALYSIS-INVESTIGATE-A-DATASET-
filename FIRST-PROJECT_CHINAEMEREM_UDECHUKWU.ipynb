# QUESTION 2: WHAT ARE THE TOP 10 MOVIES(original_title) BY POPULARITY

# CORRELATION BETWEEN COLUMNS OF THE DATASETS

```python
data.corr()
```

**By importing seaborn, we used a heatmap to display the correlation between columns of the dataset:**

```python
import seaborn as sns
```

```python
sns.heatmap(data.corr(), annot=True)
```

> correlation values ranges from -1 to 1 <br>
> -1 : perfect negative correlation <br>
> -1 < corr < -0.7: strong negative correlation <br>
> -0.7 < corr < -0.4: fair negative correlation <br>
> -0.4 < corr < -0.1: weak negative correlation <br>
> 0: no correlation <br>
> 0.1 < corr < 0.4: weak positive correlation <br>
> 0.4 < corr < 0.7: fair positive correlation <br>
> 0.7 < corr < 1.0: strong positive correlation <br>
> 1 : perfect positive correlation

# CORRELATION BETWEEN REVENUE AND BUDGET

```python
data[['revenue', 'budget']].corr()
```

# Visualization of the correlation using heatmap

```python
sns.heatmap(data[['revenue', 'budget']].corr(), annot=True)
```

**The correlation between revenue and budget is 0.734928. this is a strong positive correlation, which means an increase in the budget will lead to an increase in the revenue and vice versa.**
